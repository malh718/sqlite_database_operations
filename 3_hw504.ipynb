```python
import pandas as pd
from sqlalchemy import create_engine
import sqlite3
```

## St Josephs Example


Create temp and local DB using SQLITE

```python
conn = sqlite3.connect('health.db')
c = conn.cursor()
```

```python
c.execute("""
            CREATE TABLE s
                (
                    billing_charge  text,
                    charge_description text,
                    insurance text,
                    NDC text,
                    price text
                );
          """)

conn.commit()
```

```python

c.execute('''
  SELECT name
  FROM sqlite_master
  WHERE type='table';
  ''')

c.fetchall()

#for value in c.fetchall():
#    print(value
```

```python


c.execute('''
  SELECT * FROM s;
''')

print(c.fetchall())
```

Insert Data

```python

sql_query = """

INSERT INTO s(
      billing_charge,
      charge_description,
      insurance,
      NDC,
      price
  )
  values (
    '1234456789',
    'HC GLYCOMARK',
    '301',
    'None',
    '229.0'
  );

"""

print(sql_query)
```

```python
c.execute(sql_query)
conn.commit()
```

```python
sql_query_2 = """

select *
from s;

"""

c.execute(sql_query_2)
print(c.fetchall())
```

```python
# conn.close()

# ## create engine to connect to our sqlite DB
engine = create_engine('sqlite:///health.db')
```

```python
pd.read_sql_query("select * from s;", conn)
```

Real Data

```python
df = pd.read_excel('stjo.xlsx')


df.shape


df.columns

columnNames = list(df)
idVars = columnNames[:8]
valueVars = columnNames[8:]

stjoseph_modified = df.melt(id_vars=idVars, value_vars=valueVars)

stjoseph_modified.columns


stjoseph_modified.rename (columns={
            'Billing/ChargingCode':'billing_charge',
            'Charge Description':'charge_description',
            'Insurance_type':'insurance',
            'NDC':'NDC',
             'Price':'price'}, inplace=True)





stjoseph_modified['price'].value_counts()

stjoseph_modified
```

## Presbyterian

Create temp and local DB using SQLITE

```python
conn = sqlite3.connect('health.db')
c = conn.cursor()
```

```python
c.execute("""
            CREATE TABLE p
                (
                    Description text,
                    insurance_type text,
                    gross_charges text,
                    variable text,
                    value text
                );
          """)

conn.commit()
conn.close()
```

```python
c.execute('''
  SELECT name
  FROM sqlite_master
  WHERE type='table';
  ''')

c.fetchall()


#for value in c.fetchall():
#    print(value
```

```python


c.execute('''
  SELECT * FROM p;
''')

print(c.fetchall())
conn.commit()
conn.close()
```

Insert Data

```python

sql_query= """

INSERT INTO p (
  'Description',
  'insurance_type',
  'gross_charges',
  'variable',
  'value'
  )
  values (
    'HC IV INFUSION HYDRATION FLUIDS ADDL HR',
    '0260',
    '358.00',
    'Empire Blue Cross Blue Shield',
    '355.2'
  );



"""


print(sql_query)

conn.commit()

conn.close()
```

```python
c.executescript(sql_query)
conn.commit()
```

```python
sql_query_2 = """

select *
from p;

"""

c.execute(sql_query_2)
print(c.fetchall())
```
